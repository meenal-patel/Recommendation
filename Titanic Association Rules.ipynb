```python
df=pd.get_dummies(titanic)
df.head()
```

# Apriori Algorithm 

```python
frequent_itemsets = apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
titanic[titanic["Gender"]=="Male"].shape[0]/titanic.shape[0]
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.3)
rules
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.6).head(30)
rules
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.1)
rules[rules["lift"]>1]
```
